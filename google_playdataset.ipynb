```python
df_copy = df.copy()
```

```python
df_copy = df_copy.drop(df_copy.index[10472])
```

```python
df_copy[~df_copy['Reviews'].str.isnumeric()]
```

```python
## Convert Review Datatype to int
df_copy['Reviews'] = df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size'] = df_copy['Size'].str.replace('M', '000')
df_copy['Size'] = df_copy['Size'].str.replace('k', '')
df_copy['Size'] = df_copy['Size'].replace('Varies with device', np.nan)
```

```python
df_copy['Size'] = df_copy['Size'].apply(float)
```

```python
df_copy['Size'].isna()
```

```python
df_copy.info()
```

```python
df.head()
```

```python
df_copy['Installs'].unique()
```

```python
df['Price'].unique()
```

```python
chars_to_remove = ['+', '$', ',']
cols_to_clean=['Installs', 'Price']

for item in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols] = df_copy[cols].str.replace(item, '')
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs'] = df_copy['Installs'].astype(int)
df_copy['Price'] = df_copy['Price'].astype(float)
```

```python
df_copy.info()
```

```python
## handling last_updated feature
df_copy['Last Updated'].unique()
```

```python
df_copy['Last Updated'].isnull().sum()
```

```python
df_copy['Last Updated'] = pd.to_datetime(df_copy['Last Updated'])
df_copy['Day'] = df_copy['Last Updated'].dt.day
df_copy['Month'] = df_copy['Last Updated'].dt.month
df_copy['Year'] = df_copy['Last Updated'].dt.year
```

```python
df_copy.info()
```

```python
df_copy.to_csv('data/google_cleaned.csv')
```

Data Preprocessing

```python
numerical_features = [feature for feature in df_copy.columns if df_copy[feature].dtype != 'O']
categorical_features = [feature for feature in df_copy.columns if df_copy[feature].dtype == 'O']

## print columns
print('We have {} numerical features: {}' .format(len(numerical_features), numerical_features))
print('\nWe have {} categorical features: {}'.format(len(categorical_features), categorical_features))
```

```python
## Proportion of count data on categorical columns

for col in categorical_features:
    print(df[col].value_counts(normalize=True)*100)
    print('----------------')
```

## Data Visualization

- Proportion of count data on numerical columns

```python
plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1)

for i in range(0, len(numerical_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=df_copy[numerical_features[i]], shade=True, color='b')
    plt.xlabel(numerical_features[i])
    plt.tight_layout()
```

Observation: Rating and Year is left skewed while Reviews, Size, Installs, and Price and right skewed

- Proportion of count data on categorical columns

```python
plt.figure(figsize=(20, 15))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1)
category = ['Type', 'Content Rating']
for i in range(0, len(category)):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=df_copy[category[i]], palette='Set2')
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```

##  Most popular App Category

```python
df_copy['Category'].value_counts().plot.pie(y=df_copy['Category'], figsize=(15, 16), autopct='%1.1f')
plt.title("Most Popular App Category", fontweight='bold', fontsize='15')
```

Observations
- The most popular kinds of Apps in playstore are under category of Family, Game, and Tools
- Beauty, Comics, Arts and Weather are least popular category apps in playstore

```python
## Top 10 app categories

category = pd.DataFrame(df_copy['Category'].value_counts())
category.rename(columns= {'Category' : 'Count'}, inplace=True)
```

```python
category
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(x=category.index[:10], y ='count', data=category[:10], palette='hls')
plt.title('Top 10 App categories', fontweight='bold')
plt.xticks(rotation=45)
plt.show()
```

Insights
- Family category has the most number of apps with 18% of apps belonging to it, followed by Games category which has 11% of the apps
- Least number of apps belong to the Beauty category with less than 1% of the total apps belonging to it

```python
df_cat_installs = df_copy.groupby(['Category']) ['Installs'].sum().sort_values(ascending=False).reset_index()
df_cat_installs.Installs = df_cat_installs.Installs/1000000000 # converting into billion
df1 = df_cat_installs.head(10)

plt.figure(figsize = (15, 8))
ax = sns.barplot(data=df1, x='Installs', y='Category', palette='deep')
sns.set_context("talk")
sns.set_style("darkgrid")
ax.set_title('Category with most number of Installs', fontweight='bold', loc='center')
plt.xlabel('No. of Installions in Billions')
plt.ylabel('Category')
plt.show()
```

## Top 5 most installed apps in top Categories

```python
dfa = df_copy.groupby(['Category' ,'App'])['Installs'].sum().reset_index()
dfa = dfa.sort_values('Installs', ascending = False)
apps = ['GAME', 'COMMUNICATION', 'TOOLS', 'PRODUCTIVITY', 'SOCIAL' ]
sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):
    df2 = dfa[dfa.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= df3,x= 'Installs' ,y='App', palette='viridis')
    plt.xlabel('Installation in Millions')
    plt.ylabel('Top 5 most installed Apps')
    plt.title(app,size = 20)

plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```

# How many apps are there on Google Play Store which gets 5 ratings?

```python
ratings = df_copy.groupby(['Category', 'Installs', 'App'])['Rating'].sum().sort_values(ascending=False).reset_index()

top_rating_apps = ratings[ratings.Rating == 5.0]

print('Number of apps which gets rating of 5 are', top_rating_apps.shape[0])
top_rating_apps.head(10).sort_values(by='Installs', ascending=False)
```
